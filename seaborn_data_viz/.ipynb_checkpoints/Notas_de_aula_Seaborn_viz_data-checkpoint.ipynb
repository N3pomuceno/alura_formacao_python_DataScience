### Análise 1.1: Diferença entre final de semana e dia da Semana

```python
def dia_da_Semana(row):
    if row.dia_da_semana == "Quinta" or row.dia_da_semana == "Sexta":
        return 'Dia útil'
    else:
        return "Final de Semana"
gorjetas['fds']= gorjetas.apply(dia_da_Semana, axis=1)
```

```python
gorjetas.head()
```

```python
ax = sns.scatterplot(x='conta_total($)', y='gorjeta($)', 
                     data=gorjetas, hue='fds')
```

Scatterplot nesse caso não ajuda tanto a ter uma boa visualização da diferença entre ser no final de semana ou não. Com o boxplot deve ser melhor:

```python
ax = sns.boxplot(x='gorjeta($)', y='fds',
                     data=gorjetas)
```

```python
ax=sns.lmplot(x='conta_total($)', y='gorjeta($)',
            hue='fds', col='fds', data=gorjetas)
```

Assim podemos ter uma melhor visualização de que os valores das gorjetas tendem ser melhores no final de semana, sua causa é ainda não determinada, mais podemos ver de que forma geral os valores tendem a ser maiores. Caso especiais levam a ter valores muito altos.

### Teste de hipótese

Nossa hipótese nula H$^{null}$ pressupõe que a distribuição da gorjeta é igual nos dois grupos. Na mesma célula, adicionamos a descrição da hipótese: "A distribuição da taxa da gorjeta é a mesma nos dois grupos". Lembramos que este conteúdo foi visto no curso de Estatística II.

Em seguida, copiaremos e colaremos a sentença na sequência, substituindo null por alt para nossa hipótese alternativa H$^{alt}$. A descrição é a oposta da hipótese anterior: "A distribuição da taxa da gorjeta não é a mesma nos dois grupos".

```python
from scipy.stats import ranksums
```

```python
sobremesa = gorjetas.query("sobremesa == 'Sim'").porcentagem
```

```python
sem_sobremesa = gorjetas.query("sobremesa == 'Não'").porcentagem
```

```python
ranksums(sobremesa, sem_sobremesa)
```

```python
r = ranksums(sobremesa, sem_sobremesa)
print('O valor do p-value é {}'.format(r.pvalue.round(3)))
```

### Teste de Hipótese
Nossa hipótese nula busca descobrir se realmente a distribuição do valor da conta é igual no sábado e no domingo. Já a hipótese alternativa é justamente o contrário, e escreveremos suas descrições em células textuais da mesma forma que fizemos anteriormente.

```python
valor_conta_domingo = gorjetas.query("dia_da_semana == 'Domingo'")['conta_total($)']
```

```python
valor_conta_sabado = gorjetas.query("dia_da_semana == 'Sábado'")['conta_total($)']
```

```python
ranksums(valor_conta_domingo, valor_conta_sabado)
```

O resultado de pvalue será 0.35, ou seja, apenas aceitaremos a hipótese alternativa se este número for menor do que 0.05; como é maior, apenas nossa hipótese nula é válida. Isso significa que a distribuição do valor da conta é igual no sábado e no domingo.

Portanto, mesmo que o sábado seja mais frequentado e o domingo possua a maior média de gorjetas, não são diferentes no que diz respeito ao valor da conta.

```python
r2 = ranksums(valor_conta_domingo, valor_conta_sabado)
print('O valor do p-value é {}'.format(r2.pvalue))
```

# Aula 5: Distribuição de frequência de teste de hipóteses
### Análise 4: Hora do dia

```python
ax = sns.catplot(x='hora_do_dia', y= 'conta_total($)', 
                 data=gorjetas)
```

```python
ax = sns.catplot(x='hora_do_dia', y= 'conta_total($)', 
                 kind='swarm',data=gorjetas)
```

```python
ax=sns.violinplot(x='hora_do_dia', y= 'conta_total($)',data=gorjetas)
```

```python
ax=sns.boxplot(x='hora_do_dia', y= 'conta_total($)',data=gorjetas)
```

```python
almoco = gorjetas.query("hora_do_dia=='Almoço'")['conta_total($)']
```

```python
sns.histplot(almoco, kde=True)
```

```python
jantar = gorjetas.query("hora_do_dia=='Jantar'")['conta_total($)']
```

```python
sns.histplot(jantar, kde=True)
```

```python
gorjetas.groupby('hora_do_dia')[['conta_total($)','gorjeta($)','porcentagem']].mean()
```

A hipótese nula buscará descobrir se a distribuição é igual no almoço e no jantar enquanto a alternativa analisará se é diferente.

Portanto, chamaremos ranksums() e passaremos o valor do jantar e do almoco. Armazenaremos esta informação em uma variável r2.

```python
r2 = ranksums(jantar, almoco)
print(r2)
```

O sistema retorna um pvalue menor do que 0.05, o que significará que a distribuição é diferente nos dois períodos do dia e que a hipótese nula poderá ser rejeitada.

```python
porcentagem_almoco = gorjetas.query("hora_do_dia == 'Almoço'").porcentagem
```

```python
porcentagem_jantar = gorjetas.query("hora_do_dia == 'Jantar'").porcentagem
```

```python
ranksums(porcentagem_almoco, porcentagem_jantar)
```

```python
r3 = ranksums(porcentagem_almoco, porcentagem_jantar)
print('O valor do p-value é de {}'.format(r3.pvalue))
```

O valor do pvalue é superior a 0.05, o que significará que temos grupos parecidos. Ou seja, a distribuição da taxa da gorjeta é igual de fato no almoço e no jantar.
